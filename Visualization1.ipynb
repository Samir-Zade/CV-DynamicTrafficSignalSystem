```python
# Drop non-numeric columns if necessary
df_numeric = data.select_dtypes(include=[float, int])

correlation_matrix = df_numeric.corr()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Set the style for the plots
sns.set(style="whitegrid")

# 1. Scatter Plot with Regression Line
plt.figure(figsize=(10, 6))
sns.regplot(x='Number of Vehicles', y='Time Allocated (seconds)', data=data, scatter_kws={'s': 50}, line_kws={"color": "red"})
plt.title('Scatter Plot with Regression Line: Number of Vehicles vs Time Allocated')
plt.xlabel('Number of Vehicles')
plt.ylabel('Time Allocated (seconds)')
plt.show()

# 2. Box Plot for each variable
plt.figure(figsize=(14, 6))
sns.boxplot(data=data)
plt.title('Box Plot: Distribution of Variables')
plt.ylabel('Value')
plt.show()

# 3. Heatmap of Correlations

plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()

# 4. Bar Chart: Average Time Allocated across different ranges of Number of Vehicles
data['Vehicle Range'] = pd.cut(data['Number of Vehicles'], bins=[0, 3, 6, 9, 12, 15], labels=['1-3', '4-6', '7-9', '10-12', '13-15'])
avg_time_allocated = data.groupby('Vehicle Range')['Time Allocated (seconds)'].mean()

plt.figure(figsize=(10, 6))
avg_time_allocated.plot(kind='bar', color='skyblue')
plt.title('Average Time Allocated by Vehicle Range')
plt.xlabel('Vehicle Range')
plt.ylabel('Average Time Allocated (seconds)')
plt.show()

# 5. Pair Plot to explore pairwise relationships
sns.pairplot(data)
plt.suptitle('Pair Plot: Exploring Relationships between Variables', y=1.02)
plt.show()
```

```python
# Treemap
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from pandas.plotting import autocorrelation_plot, lag_plot, andrews_curves, parallel_coordinates
import squarify
from mpl_toolkits.mplot3d import Axes3D



# Scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(df['Number of Vehicles'], df['Time Allocated (seconds)'], c='blue', label='Allocated Time')
plt.scatter(df['Number of Vehicles'], df['Fixed Time'], c='red', label='Fixed Time')
plt.xlabel('Total Vehicles')
plt.ylabel('Time (seconds)')
plt.title('Scatter Plot of Vehicles vs Time Allocations')
plt.legend()
plt.show()

# Line plot
plt.figure(figsize=(8, 6))
plt.plot(df['Number of Vehicles'], df['Time Allocated (seconds)'], marker='o', linestyle='-', color='blue', label='Allocated Time')
plt.plot(df['Number of Vehicles'], df['Fixed Time'], marker='x', linestyle='--', color='red', label='Fixed Time')
plt.xlabel('Total Vehicles')
plt.ylabel('Time (seconds)')
plt.title('Line Plot of Vehicles vs Time Allocations')
plt.legend()
plt.show()

# Histogram
plt.figure(figsize=(8, 6))
plt.hist(df['Time Allocated (seconds)'], bins=15, alpha=0.7, label='Allocated Time', color='blue')
plt.hist(df['Fixed Time'], bins=15, alpha=0.7, label='Fixed Time', color='red')
plt.xlabel('Time (seconds)')
plt.ylabel('Frequency')
plt.title('Histogram of Allocated and Fixed Times')
plt.legend()
plt.show()

# Box plot
plt.figure(figsize=(8, 6))
sns.boxplot(data=df[['Time Allocated (seconds)', 'Fixed Time']])
plt.xlabel('Time Allocation Type')
plt.ylabel('Time (seconds)')
plt.title('Box Plot of Allocated and Fixed Times')
plt.show()

# KDE Plot
plt.figure(figsize=(8, 6))
sns.kdeplot(df['Time Allocated (seconds)'], label='Allocated Time', shade=True, color='blue')
sns.kdeplot(df['Fixed Time'], label='Fixed Time', shade=True, color='red')
plt.xlabel('Time (seconds)')
plt.ylabel('Density')
plt.title('KDE Plot of Allocated and Fixed Times')
plt.legend()
plt.show()

# Pair plot
sns.pairplot(df)
plt.suptitle('Pair Plot of Vehicles, Allocated Time, and Fixed Time', y=1.02)
plt.show()

# Correlation heatmap
df_numeric = df.select_dtypes(include=[float, int])
plt.figure(figsize=(8, 6))
sns.heatmap(df_numeric.corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()

# Violin plot
plt.figure(figsize=(8, 6))
sns.violinplot(data=df[['Time Allocated (seconds)', 'Fixed Time']])
plt.xlabel('Time Allocation Type')
plt.ylabel('Time (seconds)')
plt.title('Violin Plot of Allocated and Fixed Times')
plt.show()

# 3D Scatter plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(df['Number of Vehicles'], df['Time Allocated (seconds)'], df['Fixed Time'], c='green', marker='o')
ax.set_xlabel('Total Vehicles')
ax.set_ylabel('Allocated Time')
ax.set_zlabel('Fixed Time')
ax.set_title('3D Scatter Plot of Vehicles, Allocated Time, and Fixed Time')
plt.show()

# Joint plot
sns.jointplot(x='Number of Vehicles', y='Time Allocated (seconds)', data=df, kind='hex', color='blue')
sns.jointplot(x='Number of Vehicles', y='Fixed Time', data=df, kind='hex', color='red')
plt.show()

# Autocorrelation plot
plt.figure(figsize=(8, 6))
autocorrelation_plot(df['Time Allocated (seconds)'])
plt.title('Autocorrelation Plot of Allocated Time')
plt.show()

autocorrelation_plot(df['Fixed Time'])
plt.title('Autocorrelation Plot of Fixed Time')
plt.show()

# Lag plot
plt.figure(figsize=(8, 6))
lag_plot(df['Time Allocated (seconds)'])
plt.title('Lag Plot of Allocated Time')
plt.show()

lag_plot(df['Fixed Time'])
plt.title('Lag Plot of Fixed Time')
plt.show()

# Andrews Curves
plt.figure(figsize=(8, 6))
andrews_curves(df, 'Number of Vehicles')
plt.title('Andrews Curves of Vehicles')
plt.show()

# Parallel Coordinates
plt.figure(figsize=(8, 6))
parallel_coordinates(df, 'Number of Vehicles')
plt.title('Parallel Coordinates of Vehicles, Allocated, and Fixed Times')
plt.show()

# Ridge plot (with seaborn's kdeplot)
plt.figure(figsize=(8, 6))
sns.kdeplot(df['Time Allocated (seconds)'], label='Allocated Time', fill=True, alpha=0.3, color='blue')
sns.kdeplot(df['Fixed Time'], label='Fixed Time', fill=True, alpha=0.3, color='red')
plt.title('Ridge Plot of Allocated and Fixed Times')
plt.legend()
plt.show()

# Joint plot with regression
sns.jointplot(x='Number of Vehicles', y='Time Allocated (seconds)', data=df, kind='reg', color='blue')
sns.jointplot(x='Number of Vehicles', y='Fixed Time', data=df, kind='reg', color='red')
plt.show()

# 2D Density Plot
plt.figure(figsize=(8, 6))
sns.kdeplot(x=df['Number of Vehicles'], y=df['Time Allocated (seconds)'], cmap='Blues', shade=True, thresh=0.05)
sns.kdeplot(x=df['Number of Vehicles'], y=df['Fixed Time'], cmap='Reds', shade=True, thresh=0.05)
plt.title('2D Density Plot of Vehicles, Allocated Time, and Fixed Time')
plt.xlabel('Total Vehicles')
plt.ylabel('Time (seconds)')
plt.show()
```
